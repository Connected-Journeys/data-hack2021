```python
import pandas as pd
import openpyxl
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
```

### 1. Stemming

```python
df=pd.read_excel('../data/nodes.xlsx',sheet_name=0,engine='openpyxl')
df
```

```python
stemmer = PorterStemmer()
```

```python
stemmed_words = [stemmer.stem(word) for word in df['label']]
stemmed_words
```

```python
df['stemmed_version']=stemmed_words
```

```python
import nltk
nltk.download('wordnet')
```

### 2.Lemmatizing

```python
lemmatizer = WordNetLemmatizer()
lemmatized_words = [lemmatizer.lemmatize(word) for word in df['label']]
lemmatized_words
```
